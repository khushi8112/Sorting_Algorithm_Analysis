```python
def bubbleSort(l,n):
    for i in range(n-1):
        for j in range(n-i-1):
            if l[j]>l[j+1]:
                temp=l[j]
                l[j]=l[j+1]
                l[j+1]=temp
    return l

def selectionSort(l,n):
    for i in range(n):
        min_i=i
        for j in range(i+1,n):
            if l[j]<l[min_i]:
                min_i=j
        temp=l[i]
        l[i]=l[min_i]
        l[min_i]=temp
    return l

def insertionSort(l,n):
    for i in range(1,n):
        key = l[i]
        j=i-1
        while j>=0 and key<l[j]:
            l[j+1]=l[j]
            j-=1
        l[j+1]=key
    return l

def partition(arr,low,high):
    pivot=arr[high]
    i=low-1
    for j in range(low,high):
        if arr[j]<=pivot:
            i=i+1
            temp=arr[i]
            arr[i]=arr[j]
            arr[j]=temp
    temp=arr[i+1]
    arr[i+1]=pivot
    arr[high]=temp
    return i+1
def quick(arr,low,high):
    if low<high:
        pi=partition(arr,low,high)
        quick(arr,low,pi-1)
        quick(arr,pi+1,high)

def mergeSort(arr):
    if len(arr)>1:
        mid=len(arr)//2
        l1=arr[:mid]
        l2=arr[mid:]
        mergeSort(l1)
        mergeSort(l2)
        i=j=k=0
        while i<len(l1) and j<len(l2):
            if l1[i]<l2[j]:
                arr[k]=l1[i]
                i+=1
            else:
                arr[k]=l2[j]
                j+=1 
            k+=1
        while i<len(l1):
            arr[k]=l1[i]
            i+=1
            k+=1
        while j<len(l2):
            arr[k]=l2[j]
            j+=1
            k+=1           
            
import random,time

d={
    "bubble":[0]*5,
    "selection":[0]*5,
    "Insertion":[0]*5,
    "Quick":[0]*5,
    "Merge":[0]*5
}
ele=10
for i in range(5):
    
    l=[0]*ele
    for j in range(ele):
        num=random.randint(1,ele)
        l[j]=num
    copies=[l.copy(),l.copy(),l.copy(),l.copy(),l.copy()]
    
    start=time.time()
    bubbleSort(copies[4],len(l))
    end=time.time()
    d["bubble"][i]=end-start

    
    start=time.time()
    selectionSort(copies[0],len(l))
    end=time.time()
    d["selection"][i]=end-start

   
    start=time.time()
    insertionSort(copies[1],len(l))
    end=time.time()
    d["Insertion"][i]=end-start

    
    start=time.time()
    quick(copies[2],0,len(l)-1)
    end=time.time()
    d["Quick"][i]=end-start

   
    start=time.time()
    mergeSort(copies[3])
    end=time.time()
    d["Merge"][i]=end-start

    ele=ele*2
for i in d.values():
    print(i)




```
